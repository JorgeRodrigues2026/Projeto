## Lidando com dados faltantes

```python
# Criando um dicionário com valores nulos usando np.nan
dados_com_nulos = {
    'AlunoID': [101, 102, 103, 104, 105, 106],
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo', 'Fernanda'],
    'Nota_Matematica': [8.5, np.nan, 7.0, np.nan, 9.2, 6.5],
    'Faltas': [2, 0, np.nan, 4, np.nan, 1],
    'Cidade': ['São Paulo', 'Rio', 'Curitiba', np.nan, 'São Paulo', 'Natal']
}

df = pd.DataFrame(dados_com_nulos)
```

```python
df
```

## Agrupando dados

Agrupar será fundamental para gerar relatórios e métricas consolidadas.

Exemplo: média de preço por categoria ou total de estoque por tipo de produto.

```python
# Pra mostrar o teste que eu quero, vou criar uma coluna de categoria
data_frame['categoria'] = ['perifericos', 'perifericos', 'perifericos', 'cadeiras', 'monitores']
```

```python
data_frame
```

```python
data_frame.groupby('categoria')['valor'].min()

# dict(data_frame.groupby('categoria')['valor'].mean())
```

```python
# Eu prefiro usar o groupyby dessa forma apesar de um pouco maior
data_frame.groupby(['categoria','faixa_preco'])['valor'].mean().reset_index(name='valor_medio_por_categoria')
```

**groupby(['categoria'])** Agrupa as linhas.


**['valor']** Filtra apenas a coluna numérica de interesse (evitando erros com colunas de texto).

**.mean()** Calcula a média, gerando uma Series indexada pelas categorias.

**.reset_index(name='...')** Transforma o índice de volta em coluna e batiza a coluna calculada com o nome que você escolheu.

```python
# Usando uma estrutura de agregação explícita
data_frame.groupby(['categoria']).agg(valor_medio_por_categoria=('valor', 'max')).reset_index()
```
